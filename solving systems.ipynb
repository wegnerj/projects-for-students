```
print('This program will solve a system of two equations in two variables in the form:\n')
print('Ax + By = C')
print('Dx + Ey = F\n')

A = input("Please enter a value for A:\n")
B = input("Please enter a value for B:\n")
C = input("Please enter a value for C:\n")
D = input("Please enter a value for D:\n")
E = input("Please enter a value for E:\n")
F = input("Please enter a value for F:\n")

A = float(A)
B = float(B)
C = float(C)
D = float(D)
E = float(E)
F = float(F)

print("\nYour system of equations is:")
print(f'{A}x + {B}y = {C}')
print(f'{D}x + {E}y = {F}')

y=((C*D-F*A)/(B*D-E*A))
x=(C-B*y)/A

print(f'\nand the solution is: ({x},{y})')
```
